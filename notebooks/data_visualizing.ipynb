```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
file_path_listing = "C:/Users/asb00/OneDrive/Documents/Courses/UAB/Masters/CS 662 - NLP/project/Listing_Titles.tsv"
file_path_test = "C:/Users/asb00/OneDrive/Documents/Courses/UAB/Masters/CS 662 - NLP/project/test.tsv"
 
df = pd.read_csv(file_path_test, sep='\t')

title_list = df['Title']
record_number_list = df['Record Number']


df['tokenized_sent'] = df.apply(lambda r: word_tokenize(r['Title']), axis=1)

print(df)
```

```python
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns
```

```python

flat_tokens = [token for sentence_tokens in df['tokenized_sent'] for token in sentence_tokens]


text = ' '.join(flat_tokens)

# Tokenize the text into words
word_counts = Counter(text)

plt.figure(figsize=(10, len(word_counts) / 3))
sns.barplot(x=list(word_counts.values()), y=list(word_counts.keys()), color='skyblue')
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.title('Word Frequency in Text')
plt.show()
```
