```python
text = ' '.join(df['Title'])

print(f"Length of the text data in dataset is {len(text)}.")
```

```python
def show_word_cloud(data, size):
    """
    Generate and display a word cloud visualization.

    Parameters:
    - data (str): Text data for generating the word cloud.
    - size (tuple): Size of the word cloud figure (width, height).

    Returns:
    None
    """
    wc = WordCloud(width=size[0], height=size[1]).generate(text=data)

    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.show()

show_word_cloud(data=text, size=(1600, 1200))
```

```python
total_tokens = ' '.join(str(df['Token']))

print(f"Total length of the tokens in dataset is {len(total_tokens)}.")
```

```python
show_word_cloud(total_tokens, size=(1600, 1200))
```
